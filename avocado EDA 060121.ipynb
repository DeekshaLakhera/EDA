# Transformation

```python
import scipy.stats as stat
import pylab
```

```python
# QQ plot
def plot(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
for i in df_num.columns:
    print(i)
    plot(df,i)
```

```python
df['Total Volume']=np.log(df['Total Volume'])
```

```python
def boxcox(feature):
    df[feature],parameters=stat.boxcox(df[feature]+1)
```

```python
for i in df_num.columns:
    if i!='year':
        print(i)
        boxcox(i)
        plot(df,i)
```

```python
for i in df_num.columns:
    print(i)
    boxplot(df,i)
    plt.show()
```
